## Part 1: Type Conversion üîÑ

Every piece of data in Python has a **type**. The three most common types are:

| Type | Description | Examples |
|------|-------------|----------|
| `int` | Whole numbers | `42`, `-7`, `0` |
| `float` | Decimal numbers | `3.14`, `-0.5`, `2.0` |
| `str` | Text (strings) | `"hello"`, `"42"`, `"3.14"` |

### Checking Types with `type()`

You can check the type of any value using the `type()` function:

````python
# Check the type of different values
print(type(42))       # Integer
print(type(3.14))     # Float
print(type("hello"))  # String
````

````python
# IMPORTANT: Numbers in quotes are strings, not numbers!
print(type(42))    # This is an integer
print(type("42"))  # This is a STRING (text that looks like a number)
````

### üö® The Problem: Mixing Types

What happens when you try to add a string and a number together?

````python
# This will cause an ERROR!
age = "17"
print(age + 5)
````

Python can't add a string to a number - it doesn't know what you want!

And watch out for this sneaky behaviour:

````python
# What does + do with two strings?
first = "17"
second = "5"
result = first + second
print(result)  # Is this 22? Or something else?
````

When you use `+` with strings, Python **concatenates** (joins) them together instead of adding them!

### ‚úÖ The Solution: Type Conversion Functions

Python gives us three functions to convert between types:

| Function | What it does | Example |
|----------|-------------|----------|
| `int()` | Converts to integer | `int("42")` ‚Üí `42` |
| `float()` | Converts to decimal | `float("3.14")` ‚Üí `3.14` |
| `str()` | Converts to string | `str(42)` ‚Üí `"42"` |

````python
# Convert string to integer
age_text = "17"
age_number = int(age_text)
print(age_number + 5)  # Now it works! Output: 22
````

````python
# Convert string to float
price_text = "19.99"
price_number = float(price_text)
print(price_number * 2)  # Output: 39.98
````

````python
# Convert number to string (useful for joining with text)
score = 85
message = "Your score is " + str(score) + " points"
print(message)
````

## Part 3: For Loops üîÅ

### The Problem: Repetitive Code

What if we want to print every student's name from our gaming survey? 

The tedious way would be:

````python
# The hard way - printing each student one by one
print(students[0])
print(students[1])
print(students[2])
print(students[3])
print(students[4])
print(students[5])

# What if we had 100 students? 1000?
````

### The Solution: For Loops! üéâ

A **for loop** automatically goes through each item in a list:

```python
for item in my_list:
    do_something_with(item)
```

The anatomy of a for loop:
- `for` - keyword that starts the loop
- `item` - a variable that holds the current item (you choose the name)
- `in` - keyword that means "from"
- `my_list` - the list to loop through
- `:` - colon at the end of the line
- **Indented code** - runs once for each item

````python
# The smart way - using a for loop
for student in students:
    print(student)
````

````python
# We can do more than just print!
for student in students:
    print(f"Welcome to the gaming survey, {student}!")
````

### Looping Through Multiple Lists Together

What if we want to print both the student name AND their gaming hours?

We use `range(len(list))` to get index numbers, then use those to access multiple lists:

````python
# Understanding range(len(list))
print(f"We have {len(students)} students")
print(f"range(6) gives us: 0, 1, 2, 3, 4, 5")
print("")

# Use the index to access multiple lists
for i in range(len(students)):
    print(f"{students[i]} plays {hours_per_week[i]} hours per week")
````

### Combining Loops with Conditionals

Now the real power - we can make decisions inside our loops!

Let's categorise gamers based on their weekly hours:

````python
# Categorise gamers based on hours
print("Gaming Categories:")
print("-" * 30)

for i in range(len(students)):
    if hours_per_week[i] >= 20:
        category = "Hardcore Gamer üî•"
    elif hours_per_week[i] >= 10:
        category = "Regular Gamer üéÆ"
    else:
        category = "Casual Gamer üòä"
    
    print(f"{students[i]}: {hours_per_week[i]} hrs/week ‚Üí {category}")
````

## Part 4: The `range()` Function üî¢

`range()` generates a sequence of numbers. It has three forms:

| Form | What it generates | Example output |
|------|-------------------|----------------|
| `range(5)` | 0 to 4 | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1 to 5 | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0 to 8, step 2 | 0, 2, 4, 6, 8 |

````python
# range(stop) - starts at 0, goes up to (but not including) stop
print("range(5):")
for num in range(5):
    print(num, end=" ")
````

````python
# range(start, stop) - starts at start, goes up to (but not including) stop
print("range(1, 6):")
for num in range(1, 6):
    print(num, end=" ")
````

````python
# range(start, stop, step) - with custom step size
print("range(0, 10, 2):")
for num in range(0, 10, 2):
    print(num, end=" ")
````

### Multiplication Tables: The Hard Way vs The Smart Way

Remember in Lesson 3 we used f-strings to format output? Now let's combine f-strings with loops!

**The tedious way** (don't do this!):

````python
# The HARD way - writing each line by hand üò´
number = 7
print(f"{number} x 1 = {number * 1}")
print(f"{number} x 2 = {number * 2}")
print(f"{number} x 3 = {number * 3}")
# ... imagine writing 10 more lines!
````

````python
# The SMART way - using a for loop! üéâ
number = 7
print(f"The {number} times table:")
print("-" * 15)

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
````

### Why Loops + F-Strings Are Powerful

| Aspect | Hard Way | Smart Way |
|--------|----------|----------|
| Lines of code | 10 | 3 |
| Change to 1-20 | Write 10 more lines | Change `11` to `21` |
| Change number | Edit 10 places | Edit 1 place |
| Risk of typos | High | Low |

### Interactive Multiplication Table

Now let's combine **user input** with **loops** and **f-strings**:

````python
# Interactive times table generator
number = int(input("Which times table do you want? "))
limit = int(input("Go up to what number? "))

print(f"\nThe {number} times table (up to {limit}):")
print("-" * 20)

for i in range(1, limit + 1):
    print(f"{number} x {i} = {number * i}")
````

## Part 5: Bar Charts üìä

Last lesson you learned about **pie charts** which show parts of a whole.

**Bar charts** are different - they compare values across categories side by side.

| Use Pie Chart When... | Use Bar Chart When... |
|----------------------|----------------------|
| Showing percentages | Comparing values |
| Parts add up to 100% | Categories are independent |
| 2-6 categories max | Any number of categories |

Let's visualise our gaming survey data!

````python
# Import matplotlib
import matplotlib.pyplot as plt

# Basic bar chart
plt.bar(students, hours_per_week)
plt.title("Weekly Gaming Hours by Student")
plt.xlabel("Student")
plt.ylabel("Hours per Week")
plt.show()
````

### Customising Bar Charts with Colours

We can use a loop to assign different colours based on conditions!

````python
# Create a colour list based on gaming hours
colors = []

for hours in hours_per_week:
    if hours >= 20:
        colors.append("red")      # Hardcore gamer
    elif hours >= 10:
        colors.append("orange")   # Regular gamer
    else:
        colors.append("green")    # Casual gamer

# Create the bar chart with colours
plt.figure(figsize=(10, 6))
plt.bar(students, hours_per_week, color=colors)
plt.title("Weekly Gaming Hours by Student")
plt.xlabel("Student")
plt.ylabel("Hours per Week")

# Add a horizontal line showing "recommended" limit
plt.axhline(y=15, color='blue', linestyle='--', label='Recommended limit')
plt.legend()

plt.show()

print("Legend: üî¥ Red = 20+ hrs | üü† Orange = 10-19 hrs | üü¢ Green = Under 10 hrs")
````

### Bar Chart Reference

| Function | What it does |
|----------|-------------|
| `plt.bar(x, y)` | Vertical bar chart |
| `plt.barh(x, y)` | Horizontal bar chart |
| `plt.axhline(y=value)` | Add horizontal reference line |
| `plt.xticks(rotation=45)` | Rotate x-axis labels |

````python
# Horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(students, hours_per_week, color=colors)
plt.title("Weekly Gaming Hours by Student")
plt.xlabel("Hours per Week")
plt.ylabel("Student")
plt.show()
````

## ‚ö†Ô∏è Common Mistakes

| Mistake | Example | Fix |
|---------|---------|-----|
| Not converting input | `age = input("Age: ")` | `age = int(input("Age: "))` |
| Forgetting colon | `for i in range(5)` | `for i in range(5):` |
| Wrong indentation | Code runs once | Indent code inside loop |
| Off-by-one error | `range(1,5)` gives 1-4 | `range(1,6)` for 1-5 |

````python
# Common mistake: Forgetting to indent
# This prints ALL numbers, then "Done!" once:
for i in range(3):
    print(i)
print("Done!")  # Not indented - runs AFTER loop finishes

print("---")

# This prints "Done!" after EACH number:
for i in range(3):
    print(i)
    print("Done!")  # Indented - runs INSIDE loop
````
