## Part 0: F-Strings â€” Formatting Text the Easy Way

Remember how we used `print()` to display text and calculations? F-strings make it much easier to combine text with variables.

### The Old Way vs The New Way

Let's say we have a player's name and score:

````python
# The old way - using commas (works, but can be awkward)
player = "Alex"
score = 2500

print("Player:", player, "scored", score, "points")
````

````python
# The NEW way - using f-strings (cleaner and more flexible!)
player = "Alex"
score = 2500

print(f"Player: {player} scored {score} points")
````

### F-String Syntax

```python
f"Your text here {variable} more text {another_variable}"
```

**Key points:**
- Start with `f` before the opening quote
- Put variables inside curly braces `{}`
- You can even put calculations inside the braces!

````python
# You can do calculations inside the curly braces!
price = 15
quantity = 4

print(f"Total cost: Â£{price * quantity}")
````

### âš ï¸ Common Mistake: `=` vs `==`

This trips up almost everyone at first!

- `=` is for **assignment** (giving a value to a variable)
- `==` is for **comparison** (checking if two things are equal)

```python
x = 5      # This ASSIGNS the value 5 to x
x == 5     # This CHECKS if x equals 5 (returns True or False)
```

````python
# Using comparisons with variables
score = 75
passing_mark = 70

print(f"Score: {score}")
print(f"Passing mark: {passing_mark}")
print(f"Did the student pass? {score >= passing_mark}")
````

## Part 2: Conditional Statements

Now that we can compare values, we can make our programs **make decisions**. This is where programming gets really powerful!

### The `if` Statement

The simplest conditional runs code only when a condition is `True`:

```python
if condition:
    # This code runs only if condition is True
    # Notice the indentation!
```

````python
# A simple if statement
age = 18

if age >= 18:
    print("You are an adult.")

print("Program finished.")  # This always runs (not indented under 'if')
````

````python
# Try changing age to 15 and run again
age = 15

if age >= 18:
    print("You are an adult.")

print("Program finished.")
````

### âš ï¸ Indentation Matters!

Python uses **indentation** (spaces at the start of a line) to know which code belongs inside the `if` block. This is the same pattern you'll see with loops in Lesson 4.

```python
if score > 50:
    print("This is inside the if")     # Indented = part of the if
    print("This is also inside")       # Still indented = still part of the if
print("This is outside the if")        # Not indented = runs regardless
```

**Tip:** Use 4 spaces for indentation. Colab usually does this automatically when you press Tab.

### `if-else`: Two-Way Decisions

What if we want to do something different when the condition is `False`? Use `else`:

````python
# Pass or fail based on score
score = 65
passing_mark = 70

if score >= passing_mark:
    print("Congratulations! You passed! ðŸŽ‰")
else:
    print("Sorry, you didn't pass this time.")
````

````python
# Change score to 75 and run again
score = 75
passing_mark = 70

if score >= passing_mark:
    print("Congratulations! You passed! ðŸŽ‰")
else:
    print("Sorry, you didn't pass this time.")
````

### `elif`: Multiple Conditions

Sometimes you need to check several conditions. That's where `elif` (short for "else if") comes in:

```python
if condition1:
    # Runs if condition1 is True
elif condition2:
    # Runs if condition1 is False AND condition2 is True
elif condition3:
    # Runs if condition1 and condition2 are False AND condition3 is True
else:
    # Runs if ALL conditions above are False
```

**Important:** Python checks conditions from top to bottom and stops at the **first** one that's `True`.

````python
# Grade calculator
score = 82

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} â†’ Grade: {grade}")
````

### ðŸŽ¯ Try It: Change the score

Run the cell above with different scores: 95, 73, 55, 100. Does the grade change as expected?

### Why Order Matters

Consider this example:

````python
# What happens with this order?
score = 95

if score >= 60:      # 95 >= 60 is True, so we stop here!
    grade = "D"
elif score >= 70:
    grade = "C"
elif score >= 80:
    grade = "B"
elif score >= 90:
    grade = "A"      # This never gets checked
else:
    grade = "F"

print(f"Score: {score} â†’ Grade: {grade}")  # Oops! Grade D for 95?
````

**Lesson:** When using `elif`, check the **most specific** condition first (highest values first when checking "greater than").

## Part 3: Pie Charts with Matplotlib

Now let's visualize data! **Pie charts** show how parts make up a whole â€” perfect for showing percentages or proportions.

### What is a Pie Chart?

A pie chart divides a circle into slices, where each slice represents a portion of the total. Here's an example:

![Sample Pie Chart](https://raw.githubusercontent.com/simunaqv2/python-jupyter-course/main/lessons/lesson03/sample_pie_chart.png)

In a pie chart:
- The whole circle represents 100% of the data
- Each slice shows what percentage that category contributes
- Larger slices = larger portions of the whole

### When to Use Pie Charts

Pie charts work best when:
- You want to show **parts of a whole** (percentages that add up to 100%)
- You have a **small number of categories** (ideally 2-6)
- The differences between slices are **easy to see**

### Our Dataset: Gaming Hours

Let's visualize how a student spends their gaming time across different games:

| Game | Hours Played | Genre | Player Rating |
|------|--------------|-------|---------------|
| Minecraft | 18 | Sandbox | 94 |
| Fortnite | 14 | Battle Royale | 89 |
| FIFA 25 | 10 | Sports | 86 |
| Roblox | 8 | Platform | 82 |

````python
# First, let's import matplotlib (you should remember this from Lesson 2!)
import matplotlib.pyplot as plt
````

````python
# Store our data in lists
games = ["Minecraft", "Fortnite", "FIFA 25", "Roblox"]
hours = [18, 14, 10, 8]

print(f"Games: {games}")
print(f"Hours: {hours}")
print(f"Total hours: {sum(hours)}")
````

### Creating a Basic Pie Chart

````python
# Create a simple pie chart
games = ["Minecraft", "Fortnite", "FIFA 25", "Roblox"]
hours = [18, 14, 10, 8]

plt.pie(hours, labels=games)
plt.title("Weekly Gaming Hours")
plt.show()
````

### Adding Percentages

It's often helpful to show the percentage each slice represents. We use `autopct`:

````python
# Pie chart with percentages
games = ["Minecraft", "Fortnite", "FIFA 25", "Roblox"]
hours = [18, 14, 10, 8]

plt.pie(hours, labels=games, autopct='%.1f%%')
plt.title("Weekly Gaming Hours")
plt.show()
````

**What does `autopct='%.1f%%'` mean?**
- `%` starts the format code
- `.1f` means show 1 decimal place
- `%%` prints an actual % symbol

Try changing `.1f` to `.0f` for whole numbers, or `.2f` for 2 decimal places!

### Customizing Colors

You can specify your own colors for the slices:

````python
# Pie chart with custom colors
games = ["Minecraft", "Fortnite", "FIFA 25", "Roblox"]
hours = [18, 14, 10, 8]
colors = ['green', 'purple', 'blue', 'red']

plt.pie(hours, labels=games, autopct='%.1f%%', colors=colors)
plt.title("Weekly Gaming Hours")
plt.show()
````

### Highlighting a Slice with `explode`

You can "pull out" a slice to emphasize it:

````python
# Explode the first slice (Minecraft)
games = ["Minecraft", "Fortnite", "FIFA 25", "Roblox"]
hours = [18, 14, 10, 8]
explode = [0.1, 0, 0, 0]  # 0.1 = pull out 10%, 0 = no pull

plt.pie(hours, labels=games, autopct='%.1f%%', explode=explode)
plt.title("Weekly Gaming Hours")
plt.show()
````
