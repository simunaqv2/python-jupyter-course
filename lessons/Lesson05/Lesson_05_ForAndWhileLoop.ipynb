                                      WHILE Loop
You don't know how many times the loop will continue. You know when when it should stop.

It is used to execute a set of instructions as long a CONDITION is true (or it stops when the condition becomes false)

while-loop-flowchart-2048x1325.png

```
numbers = [1, 2, 3, 4, 5]

i = 0
while i < len(numbers):     // CONDITION
    print(numbers[i])
    i += 1

```
Question: When will this loop become endless?

````python
numbers = [1, 2, 3, 4, 5]
print("The lenght of numbers array is", len(numbers))
i = 0
while i < len(numbers):
  print(f"The value at position {i} is:", numbers[i])
  i += 1
````

````python
import math
num = 1
while num != 0:
  numberAsString = input("Enter a number to square (Enter 0 to quit.):")
  num = int(numberAsString)
  if(num != 0):
    print("The square of the number is", math.pow(num,2))
  else:
    print("Thank you for using this program.")

````

Create a Password checker program which:
1. Accepts a password from the user by displaying "Please enter your password:".
2. Validates the password and displays "Login successful." or "Wrong password."
3. If the user enters wrong password 3 times, it displays "Sorry, your account is locked."


START

Set correct_password to "python123"
Set attempts to 0

WHILE attempts is less than 3

    Display "Please enter your password:"
    Read password from user

    IF password is equal to correct_password
        Display "Login successful."
        STOP the loop

    ELSE
        Increase attempts by 1
        Display "Wrong password."

END WHILE

IF attempts is equal to 3
    Display "Sorry, your account is locked."

END


````python
correct_password = "python123"
attempts = 0

while attempts < 3:
    password = input("Please enter your password: ")

    if password == correct_password:
        print("Login successful.")
        break
    else:
        attempts += 1
        print("Wrong password.")

if attempts == 3:
    print("Sorry, your account is locked.")
````

Rock Paper Scissors  Game
-----------------------------------

START

Create a list with: rock, paper, scissors

Set play to "yes"

Display "Welcome to Rock, Paper, Scissors!"

WHILE play is equal to "yes"

    Ask the player to choose rock, paper, or scissors

    Let the computer randomly choose rock, paper, or scissors

    Display the computer's choice

    IF player choice is the same as computer choice
        Display "It's a tie"

    ELSE IF player wins according to the rules
        Display "You win"

    ELSE
        Display "Computer wins"

    Ask player if they want to play again (yes/no)
    Save the answer in play

END WHILE

Display "Thanks for playing!"

END


````python
import random

choices = ["r", "p", "s"]
play = "y"

print(" Welcome to Rock, Paper, Scissors!")

while play == "y":
    player = input("Choose rock, paper, or scissors: (r,p,s)").lower()
    computer = random.choice(choices)

    print("Computer chose:", computer)

    if player == computer:
        print("It's a tie! ")

    elif (player == "r" and computer == "s") or \
         (player == "p" and computer == "r") or \
         (player == "s" and computer == "p"):
        print("You win! ")

    else:
        print("Computer wins! ")

    play = input("Do you want to play again? (y/n): ").lower()

print("Thanks for playing! ")
````
