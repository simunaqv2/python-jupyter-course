## Part 1: Computed Columns ➕

### The Idea

Sometimes the data you need isn't in the table — but you can **calculate** it from columns that are.

Think of a spreadsheet: if column A has prices and column B has quantities, you'd add a column C with `= A * B` for the total. pandas works the same way!

![Computed Column Diagram](https://pandas.pydata.org/docs/_images/05_newcolumn_1.svg)

### Creating a New Column with Maths

Our survey has `hours_per_week`. What about **monthly** hours? We just multiply by 4:

```python
survey["monthly_hours"] = survey["hours_per_week"] * 4
survey[["student", "hours_per_week", "monthly_hours"]].head()
```

````python
survey["monthly_hours"] = survey["hours_per_week"] * 4
survey[["student", "hours_per_week", "monthly_hours"]].head()
````

☝️ That one line created a brand new column for **all 25 students** at once. No loops needed!

### Another Example: Hours per Day

```python
survey["hours_per_day"] = survey["hours_per_week"] / 7
survey[["student", "hours_per_week", "hours_per_day"]].head()
```

````python
survey["hours_per_day"] = survey["hours_per_week"] / 7
survey[["student", "hours_per_week", "hours_per_day"]].head()
````

### Rounding Values

Those decimal numbers look messy. We can round them:

```python
survey["hours_per_day"] = round(survey["hours_per_week"] / 7, 1)
survey[["student", "hours_per_week", "hours_per_day"]].head()
```

````python
survey["hours_per_day"] = round(survey["hours_per_week"] / 7, 1)
survey[["student", "hours_per_week", "hours_per_day"]].head()
````

### Combining Two Columns

We can even use **two columns** together. Let's estimate total lifetime hours:

```python
# Estimated total hours = hours per week × 52 weeks × years playing
survey["total_hours_estimate"] = survey["hours_per_week"] * 52 * survey["years_playing"]
survey[["student", "hours_per_week", "years_playing", "total_hours_estimate"]].head()
```

````python
# Estimated total hours = hours per week × 52 weeks × years playing
survey["total_hours_estimate"] = survey["hours_per_week"] * 52 * survey["years_playing"]
survey[["student", "hours_per_week", "years_playing", "total_hours_estimate"]].head()
````

### 🧠 The Pattern

Creating a new column always follows the same shape:

```python
df["new_column"] = some calculation using df["existing_column"]
```

You can use `+`, `-`, `*`, `/` and `round()` — just like normal maths, but it applies to **every row at once**.

## Part 2: Sorting Data 🔢

### Why Sort?

Sorting helps you answer questions like:
- Who games the **most**?
- Who has the **least** experience?
- What does the **top 5** look like?

### Sorting by One Column

Use `.sort_values()` with the column name:

```python
# Sort by hours — lowest first (ascending is the default)
survey.sort_values("hours_per_week").head()
```

````python
# Sort by hours — lowest first (ascending is the default)
survey.sort_values("hours_per_week").head()
````

```python
# Sort by hours — highest first
survey.sort_values("hours_per_week", ascending=False).head()
```

````python
# Sort by hours — highest first
survey.sort_values("hours_per_week", ascending=False).head()
````

### Getting the Top N

Chain `.sort_values()` with `.head()` to get the top results:

```python
# Top 5 gamers by hours
top_5 = survey.sort_values("hours_per_week", ascending=False).head(5)
top_5[["student", "game", "hours_per_week"]]
```

````python
# Top 5 gamers by hours
top_5 = survey.sort_values("hours_per_week", ascending=False).head(5)
top_5[["student", "game", "hours_per_week"]]
````

### Sorting by Multiple Columns

What if two students have the same hours? We can sort by a **second column** to break ties:

```python
# Sort by platform (A-Z), then by hours (highest first) within each platform
survey.sort_values(["platform", "hours_per_week"], ascending=[True, False]).head(10)
```

````python
# Sort by platform (A-Z), then by hours (highest first) within each platform
survey.sort_values(["platform", "hours_per_week"], ascending=[True, False]).head(10)
````

### Sort Reference

| Code | What it does |
|------|-------------|
| `df.sort_values("col")` | Sort lowest → highest |
| `df.sort_values("col", ascending=False)` | Sort highest → lowest |
| `df.sort_values("col").head(5)` | Get the 5 smallest |
| `df.sort_values("col", ascending=False).head(5)` | Get the 5 largest |
| `df.sort_values(["col1", "col2"])` | Sort by two columns |

## Part 3: GroupBy — The Big One! 🏆

### The Question That Changes Everything

So far we've looked at **individual** students. But what if we want to ask:

> *"What is the average gaming hours **for each platform**?"*

Without pandas, you'd need to:
1. Find all PC gamers, calculate their average
2. Find all PlayStation gamers, calculate their average
3. Find all Xbox gamers, calculate their average
4. Find all Switch gamers, calculate their average

That's a lot of repetitive filtering! **GroupBy** does it all in one line.

### How GroupBy Works

Think of GroupBy as three steps:

1. **Split** — Divide the data into groups (e.g., by platform)
2. **Apply** — Do a calculation on each group (e.g., calculate the mean)
3. **Combine** — Put the results back together into a table

![GroupBy Diagram](https://pandas.pydata.org/docs/_images/06_groupby.svg)

### Your First GroupBy

Let's find the **average hours per week** for each platform:

```python
survey.groupby("platform")["hours_per_week"].mean()
```

````python
survey.groupby("platform")["hours_per_week"].mean()
````

🎉 **One line!** pandas grouped all 25 students by their platform, calculated the average for each group, and showed you the result.

Let's break down the syntax:

| Part | What it does |
|------|-------------|
| `survey.groupby("platform")` | Split the data into groups by platform |
| `["hours_per_week"]` | Pick the column we want to calculate on |
| `.mean()` | Calculate the average for each group |

### More GroupBy Calculations

You can swap `.mean()` for other calculations:

```python
# How many students on each platform?
survey.groupby("platform")["student"].count()
```

````python
# How many students on each platform?
survey.groupby("platform")["student"].count()
````

```python

```

````python
# Maximum hours per week in each genre?
survey.groupby("genre")["hours_per_week"].max()
````

```python

survey.groupby("genre")["hours_per_week"].sum()
```

````python
# Total hours across all students in each genre?
survey.groupby("genre")["hours_per_week"].sum()
````

### GroupBy Cheat Sheet

| Method | What it calculates |
|--------|-------------------|
| `.mean()` | Average |
| `.sum()` | Total |
| `.count()` | Number of items |
| `.max()` | Highest value |
| `.min()` | Lowest value |
| `.median()` | Middle value |

### Multiple Stats at Once

What if you want **several statistics** for each group? Use `.agg()` with a list:

```python
survey.groupby("platform")["hours_per_week"].agg(["mean", "max", "min", "count"])
```

````python
survey.groupby("platform")["hours_per_week"].agg(["mean", "max", "min", "count"])
````

☝️ One line just gave us a complete statistical summary for each platform. Imagine doing this manually with loops — it would take dozens of lines!

### Rounding GroupBy Results

The averages can look messy with many decimal places. Chain `.round(1)` to tidy them up:

```python
survey.groupby("genre")["hours_per_week"].mean().round(1)
```

````python
survey.groupby("genre")["hours_per_week"].mean().round(1)
````

## Part 4: Plotting Grouped Data 📈

### Bar Charts from GroupBy

GroupBy results can be plotted directly — just add `.plot(kind="bar")`:

```python
survey.groupby("platform")["hours_per_week"].mean().plot(kind="bar", color="steelblue")
plt.title("Average Gaming Hours by Platform")
plt.ylabel("Average Hours per Week")
plt.xlabel("Platform")
plt.xticks(rotation=0)
plt.show()
```

````python
survey.groupby("platform")["hours_per_week"].mean().plot(kind="bar", color="steelblue")
plt.title("Average Gaming Hours by Platform")
plt.ylabel("Average Hours per Week")
plt.xlabel("Platform")
plt.xticks(rotation=0)
plt.show()
````

### Horizontal Bar Charts

When category names are long, horizontal bars look cleaner:

```python
survey.groupby("genre")["hours_per_week"].mean().sort_values().plot(
    kind="barh", color="coral"
)
plt.title("Average Gaming Hours by Genre")
plt.xlabel("Average Hours per Week")
plt.show()
```

````python
survey.groupby("genre")["hours_per_week"].mean().sort_values().plot(
    kind="barh", color="coral"
)
plt.title("Average Gaming Hours by Genre")
plt.xlabel("Average Hours per Week")
plt.show()
````

☝️ Notice we chained `.sort_values()` before `.plot()` — this sorts the bars from smallest to largest, making the chart easier to read.

The real power comes when you **combine** computed columns, sorting, and groupby together.

```python

survey["monthly_hours"] = survey["hours_per_week"] * 4

survey.groupby("genre")["monthly_hours"].mean().round(1)
```

### Example: Top 3 Gamers per Platform

Sort first, then filter:


## ⚠️ Common Mistakes

| Mistake | Example | Fix |
|---------|---------|-----|
| Forgetting quotes around column name | `survey.groupby(platform)` | `survey.groupby("platform")` |
| Wrong brackets for new column | `survey("new_col") = ...` | `survey["new_col"] = ...` |
| Forgetting to select a column after groupby | `survey.groupby("genre").mean()` | `survey.groupby("genre")["hours_per_week"].mean()` |
| Plotting before calculating | `survey.groupby("genre").plot()` | `survey.groupby("genre")["col"].mean().plot()` |

---

## 📚 Key Vocabulary

| Term | Definition |
|------|------------|
| **Computed column** | A new column created by doing maths on existing columns |
| **Sort** | Arranging rows in order from lowest to highest (or vice versa) |
| **GroupBy** | Splitting data into groups, applying a calculation, and combining the results |
| **Aggregation** | A calculation that summarises many values into one (like mean, sum, count) |
| **`.agg()`** | A method that lets you apply multiple aggregation functions at once |
